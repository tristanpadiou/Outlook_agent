```python
# Simple working example without external dependencies
# Let's create a basic agent without MCP servers first to test the setup

simple_agent = Agent(llm)

async def test_simple_agent():
    try:
        result = await simple_agent.run('list tools')
        print("Simple agent test successful:")
        print(result.output)
    except Exception as e:
        print(f"Error with simple agent: {e}")

await test_simple_agent()
```

```python
# Replace with your actual Composio SSE URL
composio_server = MCPServerHTTP(
    url='https://mcp.composio.dev/composio/server/72e8b144-27cb-4ffa-a606-1ec4c49079f7/mcp')

agent_composio = Agent(llm, mcp_servers=[composio_server])
```

```python
async def main():
    async with agent_composio.run_mcp_servers():  
        result = await agent_composio.run('list tools')

await main()
```
